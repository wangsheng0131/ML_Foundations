# 11 matplotlib基础

```python
import numpy as np
x=np.linspace(0,10,100)
```

```python
y=np.sin(x)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
siny=y.copy()
```

```python
cosy=np.cos(x)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color='red')
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color='red',linestyle='--')
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color='red',linestyle='--')
plt.xlim(-5,15)
plt.ylim(0,1)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color='red',linestyle='--')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.show()
```

```python
plt.plot(x,siny,label='sin(x)')
plt.plot(x,cosy,color='red',linestyle='--',label='cos(x)')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to matplotlib world!")
plt.show()
```

## 散点图 Scatter Plot

```python
plt.scatter(x,siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```
